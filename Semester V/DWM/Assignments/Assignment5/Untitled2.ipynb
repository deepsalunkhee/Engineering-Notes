```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(0)

# Define the number of customers
num_customers = 200

# Generate random annual income and spending score data
annual_income = np.random.uniform(15000, 120000, num_customers)
spending_score = np.random.uniform(1, 100, num_customers)

# Create a DataFrame to store the generated data
customer_data = pd.DataFrame({
    'CustomerID': range(1, num_customers + 1),
    'AnnualIncome': annual_income,
    'SpendingScore': spending_score
})

# Save the generated data to a CSV file
customer_data.to_csv('customer_data.csv', index=False)

print("Generated customer data saved to 'customer_data.csv'.")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('customer_data.csv')

# Select the columns for clustering
selected_columns = data[['AnnualIncome', 'SpendingScore']]

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_columns)


for k in range(1,8):
        

        # Perform K-means clustering with the chosen k
        kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
        clusters = kmeans.fit_predict(scaled_data)
        
        # Add cluster labels to the DataFrame
        data['Cluster'] = clusters
        
        # Visualize the clusters
        plt.figure(figsize=(8, 6))
        for cluster in range(k):
            plt.scatter(data[data['Cluster'] == cluster]['AnnualIncome'],
                        data[data['Cluster'] == cluster]['SpendingScore'],
                        label=f'Cluster {cluster + 1}')
        
      
        # You can now analyze and interpret the clusters based on customer spending behavior and income.

```
