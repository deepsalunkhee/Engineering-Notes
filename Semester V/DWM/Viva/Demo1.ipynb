```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to visualize class frequencies
def visualize_class_frequencies(file1, file2):
    # Read the CSV files into dataframes
    data1 = pd.read_csv(file1)
    data2 = pd.read_csv(file2)

    # Count the class frequencies in each dataset
    class_counts1 = data1['class'].value_counts()
    class_counts2 = data2['class'].value_counts()

    # Create subplots for visualization
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot the frequencies for the first file
    axes[0].bar(class_counts1.index, class_counts1.values)
    axes[0].set_title('Human DNA Class Frequencies')
    axes[0].set_xlabel('Class')
    axes[0].set_ylabel('Frequency')

    # Plot the frequencies for the second file
    axes[1].bar(class_counts2.index, class_counts2.values)
    axes[1].set_title('Chimpanzee DNA Class Frequencies')
    axes[1].set_xlabel('Class')
    axes[1].set_ylabel('Frequency')

    # Display the plots
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Input file paths
    file1_path = input("Enter the file path for human DNA data (CSV format): ")
    file2_path = input("Enter the file path for chimpanzee DNA data (CSV format): ")

    # Call the function to visualize class frequencies
    visualize_class_frequencies(file1_path, file2_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to visualize class frequencies in a single line graph
def visualize_class_frequencies(file1, file2):
    # Read the CSV files into dataframes
    data1 = pd.read_csv(file1)
    data2 = pd.read_csv(file2)

    # Count the class frequencies in each dataset
    class_counts1 = data1['class'].value_counts()
    class_counts2 = data2['class'].value_counts()

    # Extract class labels and frequencies
    labels1, frequencies1 = class_counts1.index, class_counts1.values
    labels2, frequencies2 = class_counts2.index, class_counts2.values

    # Plot the frequencies in a single line graph with different colors
    plt.figure(figsize=(10, 6))
    plt.plot(labels1, frequencies1, marker='o', label='Human DNA', color='blue')
    plt.plot(labels2, frequencies2, marker='o', label='Chimpanzee DNA', color='green')
    
    plt.title('Class Frequencies in Human and Chimpanzee DNA')
    plt.xlabel('Class')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)

    # Display the plot
    plt.show()

if __name__ == "__main__":
    # Input file paths
    file1_path = input("Enter the file path for human DNA data (CSV format): ")
    file2_path = input("Enter the file path for chimpanzee DNA data (CSV format): ")

    # Call the function to visualize class frequencies in a single line graph
    visualize_class_frequencies(file1_path, file2_path)
```
